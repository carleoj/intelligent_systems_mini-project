```python
### Mental health dataset based on DASS-21
# Dataset Link: https://data.mendeley.com/datasets/br82d4xkj7/1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE

df = pd.read_csv('DASS.csv', sep=',')
df.rename(columns={
    'Q1_1': 'Age',
     'Q1_2': 'Gender',
     'Q1_3': 'Marital_Status',
     'Q1_4': 'Educational_Status',
     'Q1_5': 'Occupational_Status',
     'Q1_6': 'Sleeping_Problem',
     'Q3_1_S1': 'SQ1',
     'Q3_2_S2': 'SQ2',
     'Q3_3_S3': 'SQ3',
     'Q3_4_S4': 'SQ4',
     'Q3_5_S5': 'SQ5',
     'Q3_6_S6': 'SQ6',
     'Q3_7_S7': 'SQ7',
     'Q3_8_A1': 'AQ8', 
     'Q3_9_A2': 'AQ9',
     'Q3_10_A3': 'AQ10',
     'Q3_11_A4': 'AQ11',
     'Q3_12_A5': 'AQ12',
     'Q3_13_A6': 'AQ13',
     'Q3_14_A7': 'AQ14',
     'Q3_15_D1': 'DQ15',
     'Q3_16_D2': 'DQ16',
     'Q3_17_D3': 'DQ17',
     'Q3_18_D4': 'DQ18',
     'Q3_19_D5': 'DQ19',
     'Q3_20_D6': 'DQ20',
     'Q3_21_D7': 'DQ21',
}, inplace=True)


pd.set_option('display.max_columns', None)
df.head()
```

```python
df.shape
```

```python
anxiety_counts = df['Anxiety_Level'].value_counts().sort_index()
stress_counts = df['Stress_Level'].value_counts().sort_index()
depression_counts = df['Depression_Level'].value_counts().sort_index()

class_dist_df = pd.DataFrame({
    'Depression': depression_counts,
    'Anxiety': anxiety_counts,
    'Stress': stress_counts
})

class_dist_df.plot(kind='bar', figsize=(6, 4), color=['#A4C3F2', '#FFB347', '#FF6F61'])

plt.title('Class Distribution Before SMOTE')
plt.xlabel('Class Level (1-5)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Category')
plt.show()
```

```python
print(anxiety_counts)
print(stress_counts)
print(depression_counts)
```

```python
# Example: Adjust feature groups based on your dataset
stress_features = ['SQ1', 'SQ2', 'SQ3', 'SQ4', 'SQ5', 'SQ7']
anxiety_features = ['AQ8', 'AQ9', 'AQ10', 'AQ11', 'AQ12', 'AQ13', 'AQ14']
depression_features = ['DQ15', 'DQ16', 'DQ17', 'DQ18', 'DQ19', 'DQ21']

# SMOTE instance
smote = SMOTE(random_state=42)
scaler = MinMaxScaler()

# Apply SMOTE for Stress_Level
X_stress = scaler.fit_transform(df[stress_features])
y_stress = df['Stress_Level']
X_res_stress, y_res_stress = smote.fit_resample(X_stress, y_stress)

# Apply SMOTE for Anxiety_Level
X_anxiety = scaler.fit_transform(df[anxiety_features])
y_anxiety = df['Anxiety_Level']
X_res_anxiety, y_res_anxiety = smote.fit_resample(X_anxiety, y_anxiety)

# Apply SMOTE for Depression_Level
X_depression = scaler.fit_transform(df[depression_features])
y_depression = df['Depression_Level']
X_res_depression, y_res_depression = smote.fit_resample(X_depression, y_depression)

# Count after SMOTE
anxiety_res_counts = pd.Series(y_res_anxiety).value_counts().sort_index()
stress_res_counts = pd.Series(y_res_stress).value_counts().sort_index()
depression_res_counts = pd.Series(y_res_depression).value_counts().sort_index()

# Combine into a DataFrame
class_dist_res_df = pd.DataFrame({
    'Depression': depression_res_counts,
    'Anxiety': anxiety_res_counts,
    'Stress': stress_res_counts
})
class_dist_res_df.plot(kind='bar', figsize=(6, 4), color=['#A4C3F2', '#FFB347', '#FF6F61'])

plt.title('Class Distribution After SMOTE')
plt.xlabel('Class Level (1-5)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Category', loc='lower right')
plt.show()

# After SMOTE, all become equal to the largest class (800 each), so total rows = 800  5 = 4000 rows for each target task.
```

```python
print(anxiety_res_counts)
print(stress_res_counts)
print(depression_res_counts)
```

## Machine Learning Pipeline
* Normalization
* Resampling
* Model Training
* Cross Validation

```python
kfold = KFold(n_splits=10, random_state=7, shuffle=True)

models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "KNN": KNeighborsClassifier(),
    "Decision Trees": DecisionTreeClassifier(),
    "Extra Trees": ExtraTreesClassifier(n_estimators=100, max_features=7),
    "Bagging Decision Trees": BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, random_state=42)
}
```

```python
# Stress Level Prediction
X = df[['SQ1', 'SQ2', 'SQ3', 'SQ4', 'SQ5', 'SQ7']]
y = df['Stress_Level']

print("Average Accuracy with SMOTE:\n")
for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),  
        ('smote', SMOTE(random_state=42)),  
        ('model', model) 
    ])
    results = cross_val_score(pipeline, X, y, cv=kfold, scoring='accuracy')
    print(f"{name}: {results.mean():.4f}")
```

```python
# Anxiety Level Prediction
X = df[['AQ8', 'AQ9', 'AQ10', 'AQ11', 'AQ12', 'AQ13', 'AQ14']]
y = df['Anxiety_Level']

print("Average Accuracy with SMOTE:\n")
for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),  
        ('smote', SMOTE(random_state=42)),  
        ('model', model)  
    ])
    results = cross_val_score(pipeline, X, y, cv=kfold, scoring='accuracy')
    print(f"{name}: {results.mean():.4f}")
```

```python
# Depression Level Prediction
X = df[['DQ15', 'DQ16', 'DQ17', 'DQ18', 'DQ19', 'DQ15', 'DQ21']]
y = df['Depression_Level']

print("Average Accuracy with SMOTE:\n")
for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),  
        ('smote', SMOTE(random_state=42)),  
        ('model', model)  
    ])
    results = cross_val_score(pipeline, X, y, cv=kfold, scoring='accuracy')
    print(f"{name}: {results.mean():.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
plt.hist(df['Age'], bins=15, edgecolor='black', color=['#A4C3F2'])
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.show()

gender_counts = df['Gender'].value_counts().sort_index() 

plt.figure(figsize=(5, 4))
plt.bar([1, 2], gender_counts, edgecolor='black', color=['#A4C3F2','#FF6F61']) 
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.xticks([1, 2], ['Male', 'Female'])
plt.show()

marital_counts = df['Marital_Status'].value_counts().sort_index()  # Assumes No = 0, Yes = 1

# Labels and counts
labels = ['No', 'Yes']
counts = [marital_counts[0], marital_counts[1]]

plt.figure(figsize=(5, 4))
plt.bar(labels, counts, edgecolor='black', color=['#A4C3F2','#FF6F61'])   # You can change the colors
plt.title('Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Respondents')
plt.show()

df['EduLabel'] = df['Educational_Status'].map({
    1: 'Illiterate', 
    2: 'Primary',
    3: 'Secondary',
    4: 'High School',
    5: 'Graduation & Above'
})

order = ['Illiterate', 'Primary', 'Secondary', 'High School', 'Graduation & Above']
colors = ['lightgray', '#FFB347', '#FF6F61', 'teal', '#A4C3F2']

plt.figure(figsize=(6,4))
ax = sns.countplot(y='EduLabel', data=df, order=order, color='lightgray')  # base color

# Apply custom colors manually
for i, bar in enumerate(ax.patches):
    bar.set_facecolor(colors[i])

plt.title('Education Status Distribution')
plt.xlabel('Number of Respondents')
plt.show()
```
